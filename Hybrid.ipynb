```python
import matplotlib
matplotlib.use('TkAgg')
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import math

#Functions
def gaussian_filter(rows,columns,sigma):

    if rows%2==0:
        center_x = rows/2
    else:
        center_x = rows/2 + 1

    if columns%2==0:
        center_y = columns/2
    else:
        center_y = columns/2 + 1
 
    def gaussian(i,j):
        
        value = math.exp(-1.0 * ((i - center_x)**2 + (j - center_y)**2) / (2 * sigma**2))
        return value
 
    gaussian_array = np.zeros((rows,columns))

    for i in range(0,rows):
        for j in range(0,columns):
            gaussian_array[i][j] = gaussian(i,j)
    
    return gaussian_array


def LowPassFilter(image,sd):
    
    n,m = image.shape
    gaussian_matrix = gaussian_filter(n,m,sd)

    dft = np.fft.fft2(image)
    dftshift = np.fft.fftshift(dft)
    
    
    filterImage = dftshift * gaussian_matrix

    ifftshift = np.fft.ifftshift(filterImage)
    ifftImage = np.fft.ifft2(ifftshift)
    return ifftImage

def HighPassFilter(image,sd):
    
    n,m = image.shape
    gaussian_matrix = (1 - gaussian_filter(n,m,sd))

    dft = np.fft.fft2(image)
    dftshift = np.fft.fftshift(dft)
    filterImage = dftshift * gaussian_matrix

    ifftshift = np.fft.ifftshift(filterImage)
    ifftImage = np.fft.ifft2(ifftshift)

    return ifftImage

#Code
image1 = img.imread('Desktop\Afghan_girl_before.jpg',0) #Low Pass
image2 = img.imread('Desktop\makeup_after.jpg',0) #High Pass
if(len(image1.shape)==3):
    img_gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
if(len(image2.shape)==3):
    img_gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
alpha = 10
beta = 25
height,width=img_gray1.shape
img_gray2=cv2.resize(img_gray2,(width,height))
img1_low = LowPassFilter(img_gray1,alpha)
img2_high = HighPassFilter(img_gray2,beta)

plt.imshow(np.real(img1_low), cmap = 'gray')
plt.show()

plt.imshow(np.real(img2_high), cmap = 'gray')
plt.show()

image_hybrid = np.real(img1_low + img2_high)

plt.imshow(np.real(image_hybrid), cmap = 'gray')
plt.show()

cv2.imwrite("1.jpg",np.real(img1_low))
cv2.imwrite("2.jpg",np.real(img2_high))
cv2.imwrite("1_2.jpg",np.real(img_low))
```
